```python
class Memory():
  def __init__(self):
    self.all_test_cases=list()
    self.all_test_steps=list()


  def check_exist_testcase(self,name_tc):
    for tc in self.all_test_cases:
      if tc.name==name_tc:
        return self.all_test_cases.index(tc)
    return -1

  def check_exist_teststep(self,script):
    for ts in self.all_test_steps:
      if ts.script==script:
        return self.all_test_steps.index(ts)
    return -1

  def append_exe_his(self):
    for tc in self.all_test_cases:
      tc.exe_history.append(-1)
      tc.exe_time.append(-1)
      tc.failed_test_step.append(-1)
    for ts in self.all_test_steps:
      ts.exe_history.append({})

  def show_all_testcases(self):
    for tc in self.all_test_cases:
      print('--------------------------------------------------------------')
      tc.show_infor()

  def show_all_teststeps(self):
    for ts in self.all_test_steps:
      print('--------------------------------------------------------------')
      ts.show_infor()

  def create_element_weight_ts(self):
    for ts in self.all_test_steps:
      ts.create_weight_1()
      ts.create_weight_2()
      ts.create_weight_3()

  def show_all_element_weight_ts(self):
    for ts in self.all_test_steps:
      print(ts.script)
      print("List_element_weight_1:",ts.list_weight_1)
      print("List_element_weight_2:",ts.list_weight_2)
      print("List_element_weight_3:",ts.list_weight_3)
      print('')
      print('')

  def create_weight_ts(self,weight_type,length_of_his,cycle_ins,type_of_function):
    for ts in self.all_test_steps:
      ins_list_element_weight=ts.create_ins_list_element_weight_ts(weight_type,length_of_his,cycle_ins)
      print(ts.script)
      print(function_1(ins_list_element_weight))
      print('')
      #return function_1(ins_list_element_weight)

  def reset_before_cycle(self):
    for ts in self.all_test_steps:
      ts.reset_weight()

    for tc in self.all_test_cases:
      tc.reset_is_run()

```

```python
class Test_suite():
  def __init__(self, file_path,cycle,memory):
    self.file_path=file_path
    self.cycle=cycle
    self.test_cases=list()
    self.number_of_tests=0
    memory.append_exe_his()
    test_case_rows = {}
    with open(file_path, 'r') as report_file:
      csv_reader = csv.DictReader(report_file,delimiter=",")
      headings = next(csv_reader)
      for row in csv_reader:
        if row['Suite/Test/Step Name']=='':
          self.number_of_tests=self.number_of_tests+1
          continue
        if self.number_of_tests in test_case_rows:
          test_case_rows[self.number_of_tests].append(row)
        else:
          test_case_rows[self.number_of_tests] = [row]
    tmp_test_steps=list()
    tmp_length=0
    tmp_exe_time=0
    tmp_name=''
    tmp_verdict=''
    for test_case_index in range(1,self.number_of_tests +1):
      tmp_length=len(test_case_rows[test_case_index])-1
      tmp_name=test_case_rows[test_case_index][0]['Suite/Test/Step Name']
      tmp_exe_time=test_case_rows[test_case_index][0]['Duration']
      tmp_verdict=test_case_rows[test_case_index][0]['Status']
      check=memory.check_exist_testcase(tmp_name)
      if check==-1:
        tmp_testcase=Test_case(cycle,tmp_length,tmp_name)
        tmp_testcase.exe_history[cycle]=tmp_verdict
        tmp_testcase.exe_time[cycle]=tmp_exe_time
        memory.all_test_cases.append(tmp_testcase)
        for ts in test_case_rows[test_case_index][1:]:
          check_ts=memory.check_exist_teststep(ts['Suite/Test/Step Name'])
          if check_ts==-1:
            tmp_teststep=Test_step(ts['Suite/Test/Step Name'],cycle)
            tmp_teststep.exe_history[cycle][tmp_name]=ts['Status']
            memory.all_test_steps.append(tmp_teststep)
          else:
            tmp_teststep=memory.all_test_steps[check_ts]
            tmp_teststep.exe_history[cycle][tmp_name]=ts['Status']
          tmp_testcase.test_steps.append(tmp_teststep)
          if ts['Status']=='FAILED' and tmp_testcase.failed_test_step[cycle]==-1:
            tmp_testcase.failed_test_step[cycle]=tmp_teststep
      else:
        tmp_testcase=memory.all_test_cases[check]
        tmp_testcase.exe_history[cycle]=tmp_verdict
        tmp_testcase.exe_time[cycle]=tmp_exe_time
        for ts in test_case_rows[test_case_index][1:]:
          check_ts=memory.check_exist_teststep(ts['Suite/Test/Step Name'])
          tmp_teststep=memory.all_test_steps[check_ts]
          tmp_teststep.exe_history[cycle][tmp_name]=ts['Status']
          if ts['Status']=='FAILED' and tmp_testcase.failed_test_step[cycle]==-1:
                tmp_testcase.failed_test_step[cycle]=tmp_teststep
      self.test_cases.append(tmp_testcase)

    
  def show_infor(self):
      print("############################################################################")
      for tc in self.test_cases:
        print('*************************************')
        tc.show_infor()   

  def num_fail_test_cases(self):
      # co the tinh luc doc du lieu luon
      count = 0
      for tc in self.test_cases:
          if tc.exe_history[self.cycle] == 'FAILED':
              count += 1
      return count

  def num_distinct_faults(self):
      fail_test_steps = set()
      for tc in self.test_cases:
          if tc.failed_test_step[self.cycle] != -1:
              fail_test_steps.add(tc.failed_test_step[self.cycle])
      return len(fail_test_steps)
```

```python
class Test_case():
  def __init__(self,cycle,length,name):
    self.test_steps=list()
    self.exe_history=[-1 for i in range(cycle+1)]
    self.length=length
    self.exe_time=[-1 for i in range(cycle+1)]
    self.name=name
    self.failed_test_step=[-1 for i in range(cycle+1)]
    self.is_run=None

  def show_infor(self):
    print('Name: ',self.name)
    print('Execution history: ',self.exe_history)
    print('Length of test case: ',self.length)
    print('Executime: ',self.exe_time)
    print('All test steps:')
    for ts in self.test_steps:
      ts.print_script()
    print('Failed step',self.failed_test_step)
    print('')
    print('')

  def reset_is_run(self):
    self.is_run=False
```

```python
from typing import Type
class Test_step():
  def __init__(self,script,cycle):
    self.script=script
    self.exe_history=[{} for i in range(cycle+1)]
    self.list_weight_1=list()#s lng fail/total
    self.list_weight_2=list()#s lng test case fail
    self.list_weight_3=list()#c fail tnh l 1, nu k fail or k chy th l 0
    self.weight=-1

  def print_script(self):
    print(self.script)

  def show_infor(self):
    self.print_script()
    for cycle,tc_names in enumerate(self.exe_history):
      print("Cycle:",cycle)
      for tc_name in tc_names:
        print(tc_name,': ',self.exe_history[cycle][tc_name])
    print('')
    print('')

  def create_weight_1(self):
    for cycle,tc_names in enumerate(self.exe_history):
      if len(tc_names)==0:
        self.list_weight_1.append(0)
      else:
        number_of_failed=0
        for tc_name in tc_names:
          if self.exe_history[cycle][tc_name]=='FAILED':
            number_of_failed=number_of_failed+1
        self.list_weight_1.append(float(number_of_failed/len(tc_names)))

  def create_weight_2(self):
    for cycle,tc_names in enumerate(self.exe_history):
      if len(tc_names)==0:
        self.list_weight_2.append(0)
      else:
        number_of_failed=0
        for tc_name in tc_names:
          if self.exe_history[cycle][tc_name]=='FAILED':
            number_of_failed=number_of_failed+1
        self.list_weight_2.append(number_of_failed)

  def create_weight_3(self):
    for cycle,tc_names in enumerate(self.exe_history):
      exist_failed=0
      for tc_name in tc_names:
        if self.exe_history[cycle][tc_name]=='FAILED':
          exist_failed=1
          break
      self.list_weight_3.append(exist_failed)

  def create_ins_list_element_weight_ts(self):#cycle_ins>=1 ins_cycle
    if ins_cycle<1 or ins_cycle > len(self.list_weight_1):
      print('invalid ins_cycle')
    if type_of_element_weight==1:
      List_element_weight=self.list_weight_1
    if type_of_element_weight==2:
      List_element_weight=self.list_weight_2
    if type_of_element_weight==3:
      List_element_weight=self.list_weight_3
    tmp=List_element_weight[max(0,ins_cycle-length_vector):ins_cycle]
    while len(tmp)<length_vector:
      tmp.insert(0,0)
    
    #print(self.script)
    #print(tmp1)
    #print('')
    return tmp

  def function_1(self):
    list_element_weight=self.create_ins_list_element_weight_ts()
    final_weight=0
    for count,element_weight in enumerate(list_element_weight):
      final_weight=final_weight+(count+1)*element_weight
    return final_weight

  def function_2(self):
    list_element_weight=self.create_ins_list_element_weight_ts()
    final_weight=0
    for element_weight in list_element_weight:
      final_weight=final_weight+element_weight
    return final_weight
  
  def function_3(self):
    list_element_weight=self.create_ins_list_element_weight_ts()
    final_weight=0
    for count,element_weight in enumerate(list_element_weight):
      final_weight=final_weight+pow(gamma,length_vector-count-1)*element_weight
    return final_weight
    
  def reset_weight(self):
    self.weight=-1

```

```python
class Ranking:
    def __init__(self, test_suite):
        self.test_suite = test_suite
        self.cycle = test_suite.cycle
        self.run_test_steps = set() 

    # pass function to calculate test step
    def tc_weight_calc_3(self, test_case):
        weight = 0
        for ts in test_case.test_steps:
            if ts not in self.run_test_steps:
              weight =weight+ 1
        #test_case.show_infor()
        if len(test_case.test_steps)==0:
          weight=0
        else:
          #weight = weight/len(test_case.test_steps)
          weight = weight
        return weight


    # pass function to calculate test step
    def tc_weight_calc_2(self, test_case):
        weight = 0
        for ts in test_case.test_steps:
            if ts not in self.run_test_steps:
                if ts.weight == -1:
                    #ts.weight = ts_weight_calc(ts, self.test_suite.cycle)
                    if type_of_function=='Function 1':
                      ts.weight=ts.function_1()
                    elif type_of_function=='Function 2':
                      ts.weight=ts.function_2()
                    elif type_of_function=='Function 3':
                      ts.weight=ts.function_3()
                    #elif:
                weight =weight+ ts.weight
        #test_case.show_infor()
        if len(test_case.test_steps)==0:
          weight=0
        else:
          #weight = weight/len(test_case.test_steps)
          weight = weight
        return weight


    def tc_weight_calc_1(self,test_case):
        weight = 0
        for ts in test_case.test_steps:
            # weight of test step has not been calculated
            if ts.weight == -1:
              if type_of_function=='Function 1':
                ts.weight=ts.function_1()
              elif type_of_function=='Function 2':
                ts.weight=ts.function_2()
              elif type_of_function=='Function 3':  
                ts.weight=ts.function_3()            
            weight =weight+ ts.weight

        if len(test_case.test_steps)==0:
          weight=0
        else:
          #weight = weight/len(test_case.test_steps)
          weight = weight
        return weight

        

    def choose_next_tc(self):
        max_weight = -1
        next_tc = None
        for tc in self.test_suite.test_cases:
            if not tc.is_run: # ch ni n
                weight = self.tc_weight_calc_2(tc)
                if weight > max_weight:
                    max_weight = weight
                    next_tc = tc
        return next_tc
      
    def choose_next_tc_without_weight(self):
        max_weight = -1
        next_tc = None
        for tc in self.test_suite.test_cases:
            if not tc.is_run: # ch ni n
                weight = self.tc_weight_calc_3(tc)
                if weight > max_weight:
                    max_weight = weight
                    next_tc = tc
        return next_tc



    def ranking_2(self):
        order = []
        # O(n^2)
        for _ in range(len(self.test_suite.test_cases)):
            next_tc = self.choose_next_tc()
            order.append(next_tc)
            next_tc.is_run = True
            self.run_test_steps.update(next_tc.test_steps)
        return order

    def ranking_optimize_coverage(self):
        order = []
        # O(n^2)
        for _ in range(len(self.test_suite.test_cases)):
            next_tc = self.choose_next_tc_without_weight()
            if not next_tc is None:
              order.append(next_tc)
              next_tc.is_run = True
              self.run_test_steps.update(next_tc.test_steps)
        return order

    def ranking_1(self):
      order = sorted(self.test_suite.test_cases, key = lambda tc: self.tc_weight_calc_1(tc))
      return order

    def length_test_case(self,tc):
      return tc.length


    def ranking_length_tc(self):
      order = sorted(self.test_suite.test_cases, key = lambda tc: self.length_test_case(tc))
      return order      


    def run(self):
      if type_of_ranking=='Type 1':
        return self.ranking_1()
      elif type_of_ranking=='Type 2':
        return self.ranking_2()
```

```python
def apfd_calc_1(test_suite, order):
    num_faults = test_suite.num_fail_test_cases()
    if num_faults == 0:
        return 100 

    num_test_cases = len(order)
    tf_sum = 0

    for index in range(num_test_cases):
        test_case = order[index]
        if test_case.exe_history[test_suite.cycle] == 'FAILED': # fail
            tf_sum += (index + 1)

    apfd = (1 - tf_sum / (num_test_cases*num_faults) + 1 / (2*num_test_cases)) * 100
    return apfd

################################
def apfd_calc_2(test_suite, order):
    num_faults = test_suite.num_distinct_faults()
    if num_faults == 0:
        return 100 

    num_test_cases = len(order)
    tf_sum = 0
    fail_test_steps_found = set()

    for index in range(num_test_cases):
        test_case = order[index]
        fail_test_step = test_case.failed_test_step[test_suite.cycle]
        if fail_test_step != -1 and fail_test_step not in fail_test_steps_found:
            tf_sum += (index + 1)
            fail_test_steps_found.add(fail_test_step)

    apfd = (1 - tf_sum / (num_test_cases*num_faults) + 1 / (2*num_test_cases)) * 100
    return apfd
```

```python
def ensemble_3(order_cover,order_length,order_RLTCP,test_suite):
  ensemble_order=[]
  ensemble_scores=[]
  for test_case in test_suite.test_cases:
    ensemble_score=take_position(order_cover,test_case)+take_position(order_length,test_case)+take_position(order_RLTCP,test_case)
    smaller_elements = [element for element in ensemble_scores if element < ensemble_score]
    ensemble_order.insert(len(smaller_elements), test_case)
    ensemble_scores.append(ensemble_score)
  return ensemble_order
```

```python
def ensemble(order_1,order_2,test_suite):
  ensemble_order=[]
  ensemble_scores=[]
  for test_case in test_suite.test_cases:
    ensemble_score=take_position(order_1,test_case)+take_position(order_2,test_case)
    smaller_elements = [element for element in ensemble_scores if element < ensemble_score]
    ensemble_order.insert(len(smaller_elements), test_case)
    ensemble_scores.append(ensemble_score)
  return ensemble_order
```

```python
def ensemble_cover_RLTCP(order_cover,order_RLTCP,test_suite):
  ensemble_order=[]
  ensemble_scores=[]
  for test_case in test_suite.test_cases:
    ensemble_score=take_position(order_cover,test_case)+take_position(order_RLTCP,test_case)
    smaller_elements = [element for element in ensemble_scores if element < ensemble_score]
    ensemble_order.insert(len(smaller_elements), test_case)
    ensemble_scores.append(ensemble_score)
  return ensemble_order
```

```python
def ensemble_length_RLTCP(order_length,order_RLTCP,test_suite):
  ensemble_order=[]
  ensemble_scores=[]
  for test_case in test_suite.test_cases:
    ensemble_score=take_position(order_length,test_case)+take_position(order_RLTCP,test_case)
    smaller_elements = [element for element in ensemble_scores if element < ensemble_score]
    ensemble_order.insert(len(smaller_elements), test_case)
    ensemble_scores.append(ensemble_score)
  return ensemble_order
```

```python
def take_position(order,test_case):
  for position,tc in enumerate(order):
    if tc.name==test_case.name:
      return position
```

```python
#gamma=index+1
def function_1(ts,weight_type,length_of_his,cycle):
  list_element_weight=ts.create_ins_list_element_weight_ts(weight_type,cycle,length_of_his,cycle)
  final_weight=0
  for count,element_weight in enumerate(list_element_weight):
    final_weight=final_weight+(count+1)*element_weight
  return final_weight


def tc_weight_calc_1(test_case, cycle):
    weight = 0
    for ts in test_case.test_steps:
        # weight of test step has not been calculated
        if ts.weight == -1:
            ts.weight = ts_weight_calc(ts, cycle)
        weight =weight+ ts.weight

    weight = weight/len(test_case.test_steps)
    return weight


def ranking_1(test_suite):
    order = sorted(test_suite.test_cases, key = lambda tc: tc_weight_calc_1(tc, test_suite.cycle))
    return order

####################################################

# how to know whether a tc was executed
def choose_next_tc(test_suite, order):
    max_weight = -1
    next_tc = None
    for tc in test_suite.test_cases:
        if tc not in order: # ch ni n
            weight = tc_weight_calc_2(tc, test_suite.order)
            if weight > max_weight:
                max_weight = weight
                next_tc = tc
    return next_tc

def ranking_2(test_suite):
    run_test_steps = set() 
    order = []
    # O(n^2)
    for _ in range(test_suite.test_cases):
        next_tc = choose_next_tc(test_suite, order)
        order.append(next_tc)
        run_test_steps.update(next_tc.test_steps)
    return order

####################################################
```
